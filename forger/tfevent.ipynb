```python
import tensorflow as tf
import matplotlib.pyplot as plt

TAG_NAME = "all_losses"
tf_event_file = "F:/PythonJupyterStudy/RL_Minecraft/forger-master/forger-master/MPI_train/0902_2_52/events.out.tfevents.1630543070.ood-7.37243.123459.v2"

value_list = []
for e in tf.compat.v1.train.summary_iterator(tf_event_file):
    for v in e.summary.value:
        if v.tag == TAG_NAME:
            value = tf.make_ndarray(v.tensor)
            value_list.append(value)

plt.plot(value_list)
plt.grid()
plt.show()
```
